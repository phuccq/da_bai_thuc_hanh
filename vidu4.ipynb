### 5.6 Hàm quantile()

<font color='red' size="5" >**df.quantile(x, axis, interpolation)**</font>  
**Tính phân vị**

**interpolation = 'linear', 'lower', 'higher', 'midpoint', 'nearest'**  
**Ví dụ 1: Tính Q2 tại cột 'Price'**

```python
data.loc[:,'Price'].quantile(0.75) # x=0.75 tương đương Q3, tương đương 75%
```

### 6.1 Hàm concat()

<font color='red' size="5" >**pandas.concat([data1, data2,...], axis, join)**</font>  
**data1, data2, ... là các bảng cần gộp**  
**axis = 1 -> ghép dọc (thêm cột), axis = 0 -> ghép trên dưới (thêm hàng)**  
**join = 'outer' ->  lấy toàn bộ và lặp thì chỉ giữ 1, join = 'inner' -> chỉ lấy những cái lặp**   

```python
data
```

```python
data1 = data.loc[:,:'UmId']
data2 = data.loc[:,'UmName':]
display(data1, data2)
```

```python
data121 = pd.concat([data1, data2], axis=0, join='outer')
display(data121)
```

```python
data122 = pd.concat([data1, data2], axis=0, join='inner')
display(data122)
```

```python
data123 = pd.concat([data1, data2], axis=1, join='outer')
display(data123)
```

```python
data124 = pd.concat([data1, data2], axis=1, join='inner')
display(data124)
```

```python
data3 = data.loc[:5, :]
data4 = data.loc[6:, :]
display(data3, data4)
```

## 7. Pivot table

**Pivot table là một bảng tổng hợp dữ liệu, là công cụ hữu ích để tổng hợp, lọc và phân tích dữ liệu một cách dễ dàng, nhanh chóng.**
<font color='red' size="5" >**df.pivot_table(values=’a’, index=’b’, columns=’c’, aggfunc=‘hàm’)**</font>  
- **Kết quả tạo ra một bảng trong đó các giá trị của thuộc tính b nằm trên các hàng, các giá trị của thuộc tính c nằm trên các cột và các ô giao nhau là các giá trị của a được tính theo các hàm trong aggfunc (các ô nhận giá trị NaN nếu dữ liệu bị thiếu).** 
- **Aggfunc có thể chứa các hàm: mean, sum, max, min** 
**Ví dụ 1: Tính trung bình cộng giá của từng mặt hàng theo năm**

```python
data.pivot_table(values='Price',index='ProductName',columns='Year',aggfunc ='mean')
```

**Ví dụ 2: Tìm giá lớn nhất theo từng năm**

```python
data.pivot_table(values='Price', columns='Year',aggfunc ='max')
```

**Ví dụ 3: Tìm giá lớn nhất của từng mặt hàng**  
<font color='red' size="5" >**df.pivot_table(values=’a’, index=’b’, columns=’c’, aggfunc=‘hàm’)**</font>

```python
data5 = data.pivot_table(values='Price', index='ProductName',aggfunc ='max')
display(data5)
data5.info()
```

```python
data6 = data.pivot_table(values='Price', index='ProductId',aggfunc ='max')
display(data6)
data6.info()
```

**Ví dụ 6: Tính tổng giá và trung bình cộng số tháng bán hàng của từng mặt hàng theo địa điểm**  
<font color='red' size="5" >**df.pivot_table(values=’a’, index=’b’, columns=’c’, aggfunc=‘hàm’)**</font>

```python
data.pivot_table(values=['Price','Month'], index='ProductName', columns='Place', aggfunc ={'Price': np.sum,'Month': np.mean})
```
